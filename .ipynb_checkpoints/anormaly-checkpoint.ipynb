# 0. Simulate real signal data

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## 0.1 Pure signal data

```python
# define parameters
time=5000
x =np.arange(time)
freq = 10

# sine wave
amp=np.sin(2*np.pi*freq*(x/200))

plt.figure(figsize=(20,6))
plt.plot(x,amp)
plt.xlabel('time')
plt.ylabel('amplitude')
```

## 0.2 Noise form

```python
rd=np.random.normal(size=time,)/10

plt.figure(figsize=(20,6))
plt.plot(rd)
plt.xlabel('time')
plt.ylabel('amplitude')
```

## 0.3 Pure + Noise 

```python
sig = rd+amp

plt.figure(figsize=(20,6))
plt.plot(x,sig)
plt.xlabel('time')
plt.ylabel('amplitude')
```

## 0.4 Lose form

```python
#parameters
sigma=50
mu=100
sizeLoss=10000
#time parameter
x=np.arange(-(time-(time-mu)),(time-mu))
#x=np.arange(-100,10000)
#print(x.size)

#gausian distribution
#ga=((1/np.sqrt(2*np.pi*sigma**2))*np.exp(-(x-mu)**2/(2*sigma**2)))
gau=-((1/np.sqrt(2*np.pi*sigma**2))*np.exp((-(x-mu)**2)/(2*sigma**2)))*sizeLoss
print(gau)
print(x)
```

```python
plt.figure(figsize=(20,6))
plt.plot(x,gau)
plt.xlabel('time')
plt.ylabel('loss signal data')
```

# 0.5 Real signal data

```python
real=amp+rd+gau

plt.figure(figsize=(20,6))
plt.plot(real)
plt.xlabel('time')
plt.ylabel('loss signal data')
```
